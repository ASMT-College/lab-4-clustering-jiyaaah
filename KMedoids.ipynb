```python
import numpy as np
try:
    from sklearn_extra.cluster import KMedoids
except ImportError:
    print("Error: 'scikit-learn-extra' is not installed. Run 'conda install -c conda-forge scikit-learn-extra' in Anaconda Prompt.")
    exit(1)
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

print("Name: Jiya Hona")
print("Symbol: 29097/078\n")

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Scale the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Initialize KMedoids algorithm with 3 clusters
kmedoids = KMedoids(n_clusters=3, random_state=42)

# Fit the model
kmedoids.fit(data_scaled)

# Calculate silhouette score
sil_score = silhouette_score(data_scaled, kmedoids.labels_)
print(f"Silhouette Score: {sil_score:.2f}")

# Plot the clusters (using first two features for visualization)
plt.figure(figsize=(8, 6))
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=kmedoids.labels_, cmap='viridis', s=50, alpha=0.6, label='Data Points')
plt.scatter(data_scaled[kmedoids.medoid_indices_, 0], data_scaled[kmedoids.medoid_indices_, 1], 
            c='red', marker='x', s=200, linewidths=3, label='Medoids')
plt.title("KMedoids Clustering on Iris Dataset")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend()
plt.grid(True)
plt.show()
```
